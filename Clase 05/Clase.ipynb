```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos un array de datos
data = np.random.normal(size=1000)

```

```python
data = np.append(data, [10])

plt.hist(data, bins=50, rwidth=0.95, color='purple')
plt.show()
```

```python
plt.scatter(range(len(data)),data)
plt.show()
```

```python
# BOXPLOT

# Generamos datos
datos = np.random.normal(size=1000)

# Creamos la grafica
fig, ax = plt.subplots()
ax.boxplot(datos)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.boxplot(data)

plt.show()
```

```python
# REGLA DE LAS TRES SIGMAS O LAS TRES DESVIACIONES
# REGLA 68-95-99.7

# Generamos un conjuto de datos aleatorios
np.random.seed(42)   # Esto va a hacer que siempre que llamemos a random, nos traiga aleatorios pero siemnpre van a ser los mismos

data = np.random.normal(0,1,1000)

# Calculamos las medidas a utilizar, media o promedio y desviacion estandar std
media = np.mean(data)
desv = np.std(data)

# Calculamos los limites superior e inferior para la regla de las tres sigmas
inferior = media - 3 * desv
superior = media + 3 * desv 

# Graficamos los limites y los datos
fig, ax = plt.subplots()
ax.hist(data, bins=30, density=True, rwidth=0.95, alpha=0.5)
ax.axvline(media, color='red',linestyle='-',label='Media')
ax.axvline(inferior, color='green',linestyle='--', label='Limite inferior')
ax.axvline(superior, color='green',linestyle='--', label='Limite Superior')
ax.legend()
ax.set_xlabel('Valores')
ax.set_ylabel('Frecuencia')
plt.show()
```
